```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

num     = 1001
std     = 5 

# x  : x-coordinate data
# y1 : (noisy) y-coordinate data
# y2 : (clean) y-coordinate data 

def fun(x):
 
	# f = np.sin(x) * (1 / (1 + np.exp(-x))) 
	f = np.abs(x) * np.sin(x)

	return f

n       = np.random.rand(num)
nn      = n - np.mean(n)
x       = np.linspace(-10,10,num)
y1      = fun(x)
y2      = y1 + nn * std

plt.plot(x, y1, 'b.')
plt.show()
```

The above graph is about clean data.

```python
plt.plot(x, y2, 'k.')
plt.show()
```

The above graph is about noisy data.

```python
z=np.polyfit(x, y2, 3)
z
p=np.poly1d(z)
p9=np.poly1d(np.polyfit(x, y2, 9))
```

```python
_=plt.plot(x, y2, '.', x, p(x), '-', x, p9(x), '--')
plt.show()
```

The above graph is the polynomial curves that fit the noisy data by the least square error with varying p=0,1,....,9
